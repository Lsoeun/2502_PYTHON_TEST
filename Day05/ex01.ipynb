```python
# 10. 홀짝에 따라 다른 값 반환하기
def solution(n):
    # n이 홀수일 경우
    if n % 2 != 0:
        # 1부터 n까지 홀수들만 더해서 합을 구하고 반환
        return sum(range(1, n + 1, 2))
    
    else:
        # n이 짝수일 경우
        total = 0  # 짝수들의 제곱 합을 저장할 변수
        # 2부터 n까지 짝수들에 대해 반복
        for i in range(2, n + 1, 2): 
            total += i * i  # 짝수 i의 제곱을 total에 더함
        return total
```

```python
# 11. 조건 문자열
def solution(ineq, eq, n, m):
    answer = 0  # 기본값은 0으로 설정. 조건에 맞으면 1로 변경됩니다.
    
    # ineq가 ">"일 때 n이 m보다 커야 하므로 해당 조건을 확인
    if n > m and ineq == ">":
        answer = 1  # n이 m보다 크면 조건을 만족하므로 answer를 1로 설정

    # ineq가 "<"일 때 n이 m보다 작아야 하므로 해당 조건을 확인
    elif n < m and ineq == "<":
        answer = 1  # n이 m보다 작으면 조건을 만족하므로 answer를 1로 설정

    # n이 m과 같고, eq가 "="일 때 조건을 만족하므로 answer를 1로 설정
    elif n == m and eq == "=":
        answer = 1  # n이 m과 같고 eq가 "="이면 조건을 만족하므로 answer를 1로 설정

    return answer  # 조건에 맞으면 1, 맞지 않으면 0을 반환
```

```python
# 12. 수열과 구간 쿼리3
def solution(arr, queries):
    # queries 배열의 각 원소를 순차적으로 처리
    for query in queries:
        # query는 [i, j] 형식이므로, 이를 i, j에 각각 할당
        i, j = query

        # arr[i]와 arr[j]의 값을 서로 바꿈
        arr[i], arr[j] = arr[j], arr[i]
    
    # 모든 쿼리를 처리한 후, 수정된 arr 배열을 반환
    return arr
  
# queries는 2차원 배열. 각 원소는 [i, j] 형식으로, 배열의 인덱스 쌍을 나타냄
```

```python
# 13. 수열과 구간 쿼리
def solution(arr, queries):
    answer = []  # 결과를 저장할 리스트
    
    # 각 쿼리를 처리
    for query in queries:
        s, e, k = query  # 쿼리에서 s, e, k 값을 추출
        min_value = -1  # 처음에는 값이 없다고 가정
        
        # arr[s]부터 arr[e]까지 순차적으로 확인
        for i in range(s, e + 1):
            if arr[i] > k:  # arr[i]가 k보다 크면
                if min_value == -1 or arr[i] < min_value:  # 가장 작은 값이면
                    min_value = arr[i] # min_value 값을 교체

        answer.append(min_value)  # 쿼리 결과를 리스트에 추가

    return answer  # 모든 쿼리의 결과를 반환
```

```python
# 14. 수열과 구간 쿼리4
def solution(arr, queries):
    # 각 쿼리를 처리
    for query in queries:
        s, e, k = query  # 쿼리에서 s, e, k 값을 추출
        
        # s부터 e까지 범위에서 인덱스 i가 k의 배수이면 arr[i]에 1을 더하기
        for i in range(s, e + 1):
            if i % k == 0:  # i가 k의 배수라면
                arr[i] += 1  # arr[i]에 1을 더함
    
    return arr  # 변경된 arr 배열을 반환
```

```python
# 15. 코드 처리하기
# 주어진 문자열 code를 처음부터 한 글자씩 읽으면서 "mode"라는 상태를 변경하고, 그에 따라 특정 조건을 만족하는 글자만 골라서 새로운 문자열 ret 만들기
# mode는 0과 1 두 가지 상태가 있고, 처음에는 mode = 0으로 시작
# 문자 "1"을 만나면 mode가 바뀜(즉, "1"이 나올 때마다 mode가 0 ↔ 1로 전환)
# code를 왼쪽부터 차례대로 읽으면서 mode가 0일 때 → 짝수 번째(0, 2, 4, ...) 글자만 선택, mode가 1일 때 → 홀수 번째(1, 3, 5, ...) 글자만 선택
# 단, 문자가 "1"이면 mode를 바꾸기만 하고 선택하지 않음
def solution(code):
    mode = 0  # 초기 mode 값 설정 (0부터 시작)
    ret = ""  # 결과 문자열 초기화

    idx = 0  # 인덱스 초기화
    while idx < len(code):  # 문자열 길이만큼 반복
        if code[idx] == "1":  # 현재 문자가 "1"이면
            mode = 1 - mode  # mode를 변경 (0 → 1, 1 → 0)
        else:
            # mode에 따라 문자를 ret에 추가
            # mode가 0일 때: 짝수 인덱스만 추가
            # mode가 1일 때: 홀수 인덱스만 추가
            if (mode == 0 and idx % 2 == 0) or (mode == 1 and idx % 2 == 1):
                ret += code[idx]  # 문자 추가

        idx += 1  # 인덱스 증가

    # ret이 빈 문자열이면 "EMPTY" 반환, 아니면 ret 반환
    if ret == "":
        return "EMPTY"
    return ret
  
# 처음에 mode = 0으로 시작한다.
# 문자열 code를 왼쪽부터 하나씩 읽는다.
# "1"을 만나면 mode를 바꾼다.
# mode에 따라 짝수/홀수 위치의 문자만 선택한다.
# 최종적으로 만들어진 문자열을 반환하되, 비어 있다면 "EMPTY"를 반환한다.
```

```python
# 16. 배열 만들기2
def solution(l, r):
    result = []  # 결과를 저장할 빈 리스트
    
    # l부터 r까지 모든 수를 검사
    for num in range(l, r + 1):
        str_num = str(num)  # 현재 숫자 num을 문자열로 변환하여 각 자리를 확인할 수 있게 함
        
        is_valid = True  # 숫자가 "0"과 "5"로만 이루어졌는지 여부를 나타내는 변수
        for c in str_num:  # 숫자의 각 자리(c)를 하나씩 확인
            if c not in '05':  # 만약 해당 자리가 "0"이나 "5"가 아니면
                is_valid = False  # 유효하지 않다고 표시
                break  # 더 이상 확인할 필요가 없으므로 반복문 종료
        
        if is_valid:  # 만약 숫자가 "0"과 "5"만 포함된 경우
            result.append(num)  # 해당 숫자를 결과 리스트에 추가
    
    # 만약 조건을 만족하는 숫자가 하나도 없다면
    if len(result) == 0:
        return [-1]  # -1을 포함한 리스트를 반환
    
    return result  # 조건을 만족하는 숫자들을 반환
```

```python
# 18. 콜라츠 수열 만들기
# n이 짝수이면 n을 2로 나누고, 홀수일 경우 3 * n + 1로 바꿈

def solution(n):
    collatz_sequence = [n]  # 수열을 담을 리스트, 처음에 n을 담음
    
    while n != 1:  # n이 1이 될 때까지 반복
        if n % 2 == 0:  # n이 짝수라면
            n = n // 2  # 2로 나누기
        else:  # n이 홀수라면
            n = 3 * n + 1  # 3 * n + 1
        collatz_sequence.append(n)  # 계산된 값을 수열에 추가
    
    return collatz_sequence  # 결과적으로 완성된 수열을 반환
```

```python
# 19. 배열 만들기4
def solution(arr):
    stk = []  # 빈 배열 stk 생성
    i = 0  # 변수 i의 초기값을 0으로 설정
    
    while i < len(arr):  # i가 arr의 길이보다 작을 때까지 반복
        if stk == []:  # stk가 빈 배열이라면
            stk.append(arr[i])  # arr[i]를 stk에 추가
            i += 1  # i에 1을 더함
        elif stk[-1] < arr[i]:  # stk의 마지막 원소가 arr[i]보다 작으면
            stk.append(arr[i])  # arr[i]를 stk에 추가
            i += 1  # i에 1을 더함
        else:  # stk의 마지막 원소가 arr[i]보다 크거나 같으면
            stk.pop()  # stk의 마지막 원소를 제거
    
    return stk  # 최종적으로 stk를 반환
```

```python
# 20. 정수를 나선형으로 배치하기
def solution(n):
    # n x n 크기의 2차원 리스트를 생성 후 None으로 초기화
    answer = []
    for i in range(n):
        row = [None] * n  # 한 행을 만들고
        answer.append(row)  # 리스트에 추가

    # 방향을 나타내는 리스트 (오른쪽 → 아래 ↓ 왼쪽 ← 위 ↑ 순서)
    move = [[0, 1], [1, 0], [0, -1], [-1, 0]]
    
    x, y = 0, 0  # 현재 위치 (시작점)
    m = 0  # 현재 이동 방향을 나타내는 인덱스

    # 1부터 n**2까지 숫자를 채우기
    for i in range(1, n**2 + 1):
        answer[y][x] = i  # 현재 위치에 숫자 채우기
        # 일반적인 관례(행=y, 열=x)

        # 다음 위치 계산
        new_y, new_x = y + move[m][0], x + move[m][1]

        # 배열 범위를 벗어나거나 이미 값이 채워진 경우, 방향 변경
        # new_y < 0 : new_y가 0보다 작은 값이라면 (즉, 배열 위쪽을 벗어났다면)
        # new_y >= n : new_y가 배열의 마지막 인덱스를 넘는다면
        # new_x < 0 : new_x가 0보다 작은 값이라면 (즉, 배열 왼쪽을 벗어났다면)
        # new_x >= n : new_x가 배열의 마지막 인덱스를 넘는다면
        # answer[new_y][new_x] is not None: answer[new_y][new_x]에 값이 이미 채워져 있으면
        if new_y < 0 or new_y >= n or new_x < 0 or new_x >= n or answer[new_y][new_x] is not None:
            m = (m + 1) % 4  # 방향 변경 (오른쪽 → 아래 → 왼쪽 → 위 순환)
            # 0은 오른쪽, 1은 아래, 2는 왼쪽, 3은 위쪽을 의미 (move = [[0, 1], [1, 0], [0, -1], [-1, 0]])
            new_y, new_x = y + move[m][0], x + move[m][1]  # 방향 변경 후 다음 위치 재계산
            # new_y, new_x = y + move[m][0], x + move[m][1]: 바뀐 방향을 기준으로 새로운 (new_y, new_x)를 계산
        
        # 위치 이동
        y, x = new_y, new_x

    return answer
```

```python
# 21. 특이한 이차원 배열
# 대칭 행렬인지 확인
def solution(arr):
    n = len(arr)  # 배열의 크기 확인 (n x n)
    
    # 배열의 모든 i, j에 대해 대칭성 확인
    for i in range(n):
        for j in range(n):
            if arr[i][j] != arr[j][i]:  # 대칭이 아니라면
                return 0  # 대칭 행렬이 아니므로 0을 반환
    
    return 1  # 모든 조건을 만족하면 대칭 행렬이므로 1을 반환
```

```python
# 22. 정사각형으로 만들기
def solution(arr):
    rows = len(arr) # 행의 수
    cols = len(arr[0]) # 열의 수

    # 행이 열보다 많으면 각 행에 0을 추가하여 열의 수를 맞추기 
    # arr = [[1, 2], [3, 4], [5, 6]]
    if rows > cols:
        for row in arr:
            while len(row) < rows:  # 요소 하나의 길이(열의 길이)가 행의 길이 보다 작다면
                row.append(0) # 요소에 0을 추가. ex[1, 2, 0]
    # 열이 행보다 많으면 0으로 채워진 행을 추가하여 행의 수를 맞추기
    # arr = [[1, 2, 3], [4, 5, 6]]
    else:
        for i in range(cols - rows):  # 행의 수가 부족하면 행을 추가
            arr.append([0] * cols)

    return arr
```

```python
# 23. 이차원 배열 대각선 순회하기
def solution(board, k):
    total_sum = 0  # 합을 저장할 변수
    
    # 2차원 배열 board의 각 원소에 대해
    for i in range(len(board)):  # 행 인덱스 i
        for j in range(len(board[i])):  # 열 인덱스 j
            if i + j <= k:  # i + j가 k 이하인 경우
                total_sum += board[i][j]  # 해당 원소를 합산
    
    return total_sum  # 최종 합을 반환
```
